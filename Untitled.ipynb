```python
file_path = r'./2years_CN_func'

i = 0
for files in tqdm(os.listdir(file_path)):
    inside1 = os.path.join(file_path, files)
    for i_file in os.listdir(inside1):
        inside2 = os.path.join(inside1, i_file)
        if os.path.isdir(inside2):
            for ii_file in os.listdir(inside2):
                inside3 = os.path.join(inside2, ii_file)
                for iii_file in os.listdir(inside3):
                    inside4 = os.path.join(inside3, iii_file)
                    for i4_file in os.listdir(inside4):
                        each_file_path = os.path.join(inside4, i4_file)
                        shutil.move(each_file_path, inside1)
```

```python
import re
import nibabel as nib
# string with pattern (starting with numbers)
re_number = re.compile(r'^\d+')
# string with pattern (starting with numbers, (can be anything in middle) ending with nii)
re_number_nii = re.compile(r'^\d+.+\.nii$')
```

```python
# putting all the structural files in under one umbrella
mci_1year_path = r'./MCI (1year)'
mci_2year_path = r'./MCI (2years)'
mci_3year_path = r'./MCI (3years)'
out_path = r"./MCI_structural_all"

mci_paths = [mci_1year_path, mci_2year_path, mci_3year_path]
cnt = 0
for main_dir in mci_paths:
    for sub in tqdm(os.listdir(main_dir)):
        if not os.path.isdir(os.path.join(main_dir, sub)) or not re_number.match(sub):
            continue
        if sub.find("--") != -1:
            continue
        subject_path = os.path.join(main_dir,sub)
        for struct_func in os.listdir(subject_path):
            if not struct_func.startswith("struct"):
                continue
            struct_dir = os.path.join(subject_path, struct_func)
            cnt += 1
            for mri_file in os.listdir(struct_dir):
                if not re_number_nii.match(mri_file):
                    continue
                struct_image_path = os.path.join(struct_dir,mri_file)
                pasting_path = os.path.join(out_path, main_dir)
                pasting_path = os.path.join(pasting_path, sub)
                os.makedirs(pasting_path, exist_ok=True)
                shutil.copy(struct_image_path, pasting_path)
```

```python
# cat12 preprocessed structural files, checking their dimensions
structural_features = []
mci_1year_path = r'./MCI (1year)'
mci_2year_path = r'./MCI (2years)'
mci_3year_path = r'./MCI (3years)'
parent_path = r"./MCI_structural_all"
mri_path = "./mri"

mci_paths = [mci_1year_path, mci_2year_path, mci_3year_path]
for main_dir in mci_paths:
    year_path = os.path.join(parent_path, main_dir)
    for sub in os.listdir(year_path):
        inside_sub = os.path.join(year_path, sub)
        inside_mri = os.path.join(inside_sub, mri_path)
        image_name = "wm" + sub + ".nii"
        image_path = os.path.join(inside_mri, image_name)
        image = nib.load(image_path)
        structural_features.append(image.get_fdata())
        
import numpy as np
structural_features = np.array(structural_features)


import pickle
with open('./feature_extraction/cat12_structural_features.pkl', 'wb') as f:
    pickle.dump(structural_features, f)
```
